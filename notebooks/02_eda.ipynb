```python
import pandas as pd

df = pd.read_csv("../data/bangalore_retail_expansion_dataset.csv")

df.head()
```

```python
df.info()
df.describe()
```

```python
#population density ranking
df.sort_values(
    "population_density",
    ascending=False
).head(10)
```

```python
#income ranking
df.sort_values(
    "avg_income_lpa",
    ascending=False
)
```

```python
#competitor density
import matplotlib.pyplot as plt

df.sort_values(
    "competitor_count",
    ascending=False
).plot(
    x="locality",
    y="competitor_count",
    kind="bar"
)

plt.show()
```
